# How much crime is in Seattle and where is it located?

```python
import pandas as pd
df = pd.read_csv("Crime_Data.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Counts of all request types
df['Primary Offense Description'].value_counts().plot(kind = 'pie')
```

## Roughly one quarter of all crime reported is car theft

```python
df['Crime Subcategory'].value_counts()
```

```python
#Counts of all request types
pd.options.display.max_rows = 144
df['Primary Offense Description'].value_counts()
```

```python
thefts = df['Primary Offense Description'].str.count("THEFT").sum()

print("Total thefts:",thefts)

size = df['Primary Offense Description'].size

print("Total size:",size)

print("Total % of theft out of all crime in Seattle:", thefts/size * 100,"%")
```

```python
violent_crime = df['Crime Subcategory'].str.count("ASSAULT|ASSLT|RAPE|HOMICIDE").sum()

print("Total violent crime:", violent_crime)

murders = df['Primary Offense Description'].str.count("HOMICIDE").sum()

print("Total murders:", murders)
```
